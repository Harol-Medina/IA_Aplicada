```python
# Paso 2
lista_de_analisis = []
for numero, review in enumerate(columna_de_reviews, start=1):
    prompt = f"""
Debes analizar el sentimiento de esta resena y responder solo una palabra:
Positiva, Negativa o Neutra.

Resena: {review}
"""
    sentimiento = safe_generate(
        lambda texto: client_groq.chat.completions.create(
            model="llama-3.3-70b-versatile",
            messages=[{"role": "user", "content": texto}],
        ).choices[0].message.content,
        prompt,
        task="sentimiento",
    ).strip().split()[0]
    if sentimiento not in {"Positiva", "Negativa", "Neutra"}:
        sentimiento = "Neutra"
    lista_de_analisis.append(sentimiento)

lista_de_analisis[:10]
```

```python
#Paso 3

df_reviews["Sentimiento"] = lista_de_analisis

df_reviews.head()
```

```python
# Guardar resultado para el notebook 06
df_reviews.to_csv(OUTPUT_DIR / "reviews_sentimiento.csv", index=False)
print("Guardado en output/reviews_sentimiento.csv")
```
