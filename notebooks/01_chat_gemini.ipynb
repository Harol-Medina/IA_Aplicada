# 01 — Chat interactivo con Gemini

## Objetivo
Crear un chat de consola con el modelo **Gemini 2.5 Flash** usando la API oficial de Google.

## Conceptos
- **API Key**: credencial en `.env` como `GEMINI_API_KEY`
- **Cliente**: `genai.Client()` centraliza las llamadas
- **Sesión de chat**: mantiene contexto entre mensajes con `client.chats.create()`

## Flujo
1. Cargar la API key desde `.env`
2. Crear el cliente y la sesión de chat
3. Bucle: leer pregunta → enviar → imprimir respuesta → repetir hasta escribir `fin`

## Requisitos
- Entorno virtual `.venv` activado
- Archivo `.env` con `GEMINI_API_KEY=tu_clave`


```python
import sys
sys.path.insert(0, "..")

import pandas as pd
from notebooks._setup import (
    OUTPUT_DIR,
    LocalGeminiClient,
    load_env,
    require_key,
    safe_generate,
    save_question_answers,
    use_real_apis,
)

load_env()
OUTPUT_DIR.mkdir(exist_ok=True)

if use_real_apis():
    require_key("GEMINI_API_KEY")
    from google import genai
    client = genai.Client()
else:
    client = LocalGeminiClient()
```

```python
chat = client.chats.create(model="gemini-2.5-flash")
```

```python
preguntas = [
    "Que es la IA?",
    "Menciona tres ejemplos de uso de la IA en la vida diaria.",
]

resultados = []
for numero, pregunta in enumerate(preguntas, start=1):
    respuesta = safe_generate(lambda texto: chat.send_message(texto), pregunta)
    resultados.append({
        "numero": numero,
        "pregunta": pregunta,
        "respuesta": respuesta,
    })

save_question_answers(OUTPUT_DIR / "01_respuestas_chat_gemini.csv", resultados)

df_respuestas = pd.DataFrame(resultados)
df_respuestas
```
